```python
#1
df1 = app_train[app_train['CNT_CHILDREN'] == 0]
df2 = app_train.loc[app_train['CNT_CHILDREN'] >= 1]
df3 = app_train.loc[app_train['CNT_CHILDREN'] <= 2]
df4 = pd.concat([df2, df3], axis = 0, join = 'inner')
df5 = app_train.loc[app_train['CNT_CHILDREN'] >= 3]
df6 = app_train.loc[app_train['CNT_CHILDREN'] <= 5]
df7 = pd.concat([df5, df6], axis = 0, join = 'inner')
df8 = app_train.loc[app_train['CNT_CHILDREN'] > 5]

df1.head()
df4.head()
df7.head()
df8.head()
```

```python
#2-1
grouped_df1 = df1.groupby('TARGET')['AMT_INCOME_TOTAL']
grouped_df1.mean()

grouped_df4 = df4.groupby('TARGET')['AMT_INCOME_TOTAL']
grouped_df4.mean()

grouped_df7 = df7.groupby('TARGET')['AMT_INCOME_TOTAL']
grouped_df7.mean()

grouped_df8 = df8.groupby('TARGET')['AMT_INCOME_TOTAL']
grouped_df8.mean()
```

```python
#2-2
app_train['CNT_CHILDREN_GROUP'] = pd.concat([df1, df4, df7, df8])
plt_column = 'AMT_INCOME_TOTAL'
plt_by = ['CNT_CHILDREN_GROUP' , 'TARGET']

app_train.boxplot(column=plt_column, by = plt_by, showfliers = False, figsize=(12,12))
plt.suptitle('')
plt.show()
```
